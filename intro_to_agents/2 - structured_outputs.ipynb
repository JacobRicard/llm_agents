```python
import os

from dotenv import load_dotenv
from pydantic import BaseModel, Field

from intro_to_agents.agents.llms import OpenAILLM

# Make sure you have a .env file with your OpenAI API key
    # ex: OPENAI_API_KEY = "abcdefg12345"
    # a .env file is a text file that contains environment variables
    # it is used to store sensitive information (like API keys) in a secure way
    # it is NOT included in the repository (you don't want to share your API key with others), so you will need to create it yourself

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")
```

```python
# Instantiate the LLM
model = "gpt-5-mini" # you can change the model to whatever you want (see from intro_to_agents.agents.llms import OPENAI_TOKEN_LIMITS)
llm = OpenAILLM(api_key=api_key, model=model)
```

```python
# Define the structured output (a pydantic BaseModel) that you want to use
    # 1. define the attributes that you want to include in the structured output
    # 2. define the types of the attributes (str, float, list, etc.)
    # 3. for extra clarity, define the description of the attributes (Field(description="..."))

class Animal(BaseModel):
    name: str
    weight: float = Field(description="The weight of the animal in pounds")
    height: float = Field(description="The height of the animal in inches")
    fun_facts: list[str] = Field(description="A list of fun facts about the animal")
```

```python
animal = "octopus"
prompt = f"Please tell me more about the {animal}. I want its scientific name, its weight, its height, and 5 fun facts about it."

result = llm.structured_query(prompt = prompt, response_format = Animal)
```

```python
# Key off the output to get specific information
print(f"The scientific name of the {animal} is {result.name}.")
print(f"The weight of the {animal} is {result.weight} pounds.")
print(f"The height of the {animal} is {result.height} inches.")
print(f"Here are 5 fun facts about the {animal}:")
for fact in result.fun_facts:
    print(f"- {fact}")
```
